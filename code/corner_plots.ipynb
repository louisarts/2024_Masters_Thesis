# FLARES corner plot

```python
import numpy as np
import matplotlib.pyplot as plt
import corner
from scipy.stats import gaussian_kde
import pandas as pd

# Load the data from CSV
flares_data = pd.read_csv('../data/filtered_FLARES_data.csv')

# Extract relevant columns and convert to numpy arrays
redshifts = flares_data['Redshift'].to_numpy(dtype=float)
stellar_masses = flares_data['Stellar Mass'].to_numpy(dtype=float)
SFRS = flares_data['Star Formation Rate'].to_numpy(dtype=float)

# Mask to filter out zero SFR values
mask = SFRS != 0
nonzero_redshifts = redshifts[mask]
nonzero_stellar_masses = stellar_masses[mask]
nonzero_SFRS = np.log10(SFRS[mask])

# Combine the lists into a single dataset
data = np.vstack((nonzero_redshifts, nonzero_stellar_masses, nonzero_SFRS)).T

# Define labels and titles
labels = ["z", "$\\log_{10}(M_*/M_\\odot)$", "$\\log_{10}(SFR/M_\\odot \\ \\mathrm{yr^{-1}})$"]
titles = ["Redshift Distribution", "Stellar Mass Distribution", "SFR Distribution"]

# Generate corner plot
figure = corner.corner(data,
                       bins=30,
                       labels=labels,
                       max_n_ticks=6,
                       smooth=1,
                       plot_datapoints=True,
                       color="black",
                       use_math_text=True,
                       plot_density=True,
                       fill_contours=True)

# Access the axes from the figure
axes = np.array(figure.axes).reshape((len(labels), len(labels)))

# Set label font sizes
for ax, label in zip(axes[:, 0], labels):
    ax.set_ylabel(label, fontsize=14)
for ax, label in zip(axes[-1, :], labels):
    ax.set_xlabel(label, fontsize=14)

# Add KDE to the histograms
for i, title in enumerate(titles):
    ax = axes[i, i]
    ax.set_title(title, fontsize=12)
    
    # KDE for 1D distributions
    data_1d = data[:, i]
    kde = gaussian_kde(data_1d)
    x_vals = np.linspace(data_1d.min(), data_1d.max(), 100)
    kde_vals = kde(x_vals)
    
    # Scale KDE to match histogram
    hist_vals, _ = np.histogram(data_1d, bins=30, density=False)
    scale_factor = hist_vals.max() / kde_vals.max()
    ax.plot(x_vals, kde_vals * scale_factor, color='black', linewidth=1)

plt.show()
```

# COSMOS-Web corner plot


```python
# Path to the input CSV file
akins_csv_path = '../data/LRD_data.csv'

# Reading the CSV file
akins_data = pd.read_csv(akins_csv_path)

redshifts = akins_data["z_gal_med"].tolist()[1:]
stellar_masses = akins_data["logMstar_med"].tolist()[1:]
SFRS = akins_data["SFR_100_med"].tolist()[1:]

SFRS = list(map(float, SFRS))
SFRS = np.log10(SFRS)

# Assuming redshifts, stellar_masses, and SFRS are already defined and contain numeric data
redshifts = np.array(redshifts, dtype=float)
stellar_masses = np.array(stellar_masses, dtype=float)
SFRS = np.array(SFRS, dtype=float)

# Combine the lists into a single dataset
data = np.vstack((redshifts, stellar_masses, SFRS)).T

labels = ["z", "$\\log_{10}(M_*/M_\\odot)$", "$\\log_{10}(SFR/M_\\odot \\ \\mathrm{yr^{-1}})$"]
titles = ["Redshift Distribution", "Stellar Mass Distribution", "SFR Distribution"]

# Corner plot
figure = corner.corner(data,
                       bins=30,
                       show_titles=False,
                       labels=labels,
                       max_n_ticks=6,
                       smooth=1,
                       plot_datapoints=True,
                       color="black",
                       use_math_text=True,
                       plot_density=True,
                       fill_contours=True,
                       title_kwargs={"fontsize": 10})
axes = np.array(figure.axes).reshape((len(labels), len(labels)))

# Add KDE to the histograms
for i, title in enumerate(titles):
    ax = axes[i, i]
    ax.set_title(title, fontsize=10)
    
    # KDE
    data_1d = data[:, i]
    kde = gaussian_kde(data_1d)
    x_vals = np.linspace(data_1d.min(), data_1d.max(), 100)
    kde_vals = kde(x_vals)
    hist_vals, bin_edges = np.histogram(data_1d, bins=30, density=False)
    scale_factor = hist_vals.max() / kde_vals.max()
    kde_vals *= scale_factor
    ax.plot(x_vals, kde_vals, color='black', linewidth=1)
    
plt.show()
```
