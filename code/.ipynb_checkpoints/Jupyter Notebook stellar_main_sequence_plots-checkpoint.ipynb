```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Load the data from the CSV file
akins_csv_path = '../data/LRD_data.csv'
akins_data = pd.read_csv(akins_csv_path)

# Extract relevant columns and convert to numpy arrays
logMstar = np.array(akins_data["logMstar_med"][1:], dtype=float)
logMstar_upper68 = np.array(akins_data["logMstar_u68"][1:], dtype=float)
logMstar_lower68 = np.array(akins_data["logMstar_l68"][1:], dtype=float)
SFR_100 = np.log10(np.array(akins_data["SFR_100_med"][1:], dtype=float))
SFR_100_upper68 = np.log10(np.array(akins_data["SFR_100_u68"][1:], dtype=float))
SFR_100_lower68 = np.log10(np.array(akins_data["SFR_100_l68"][1:], dtype=float))

# Calculate uncertainties for stellar mass and SFR
logMstar_err = [logMstar - logMstar_lower68, logMstar_upper68 - logMstar]
SFR_100_err = [SFR_100 - SFR_100_lower68, SFR_100_upper68 - SFR_100]

# Identify indices for MIRI galaxies and other galaxies
miri_indices = akins_data['PB_MIRI'][1:].astype(int) == 1
other_indices = ~miri_indices  # All other galaxies that are not MIRI

# Plotting setup
plt.figure(figsize=(8, 6))

# Plot error bars for MIRI galaxies
plt.errorbar(logMstar[miri_indices], SFR_100[miri_indices], 
             xerr=[logMstar_err[0][miri_indices], logMstar_err[1][miri_indices]], 
             yerr=[SFR_100_err[0][miri_indices], SFR_100_err[1][miri_indices]], 
             fmt='none', ecolor='green', elinewidth=0.3, alpha=0.1, capsize=2, capthick=0.5, zorder=1)

# Plot error bars for other galaxies
plt.errorbar(logMstar[other_indices], SFR_100[other_indices], 
             xerr=[logMstar_err[0][other_indices], logMstar_err[1][other_indices]], 
             yerr=[SFR_100_err[0][other_indices], SFR_100_err[1][other_indices]], 
             fmt='none', ecolor='#b2182b', elinewidth=0.3, alpha=0.1, capsize=2, capthick=0.5, zorder=1)

# Scatter plot for MIRI galaxies
plt.scatter(logMstar[miri_indices], SFR_100[miri_indices], 
            c='green', s=10, alpha=0.8, linewidth=0.5, label='MIRI', zorder=2)

# Scatter plot for other galaxies
plt.scatter(logMstar[other_indices], SFR_100[other_indices], 
            c='#b2182b', s=10, alpha=0.7, linewidth=0.5, label='NIRcam', zorder=2)

# Plot specific star formation rate (sSFR) lines
log_M_star_line = np.linspace(0, 16, 100)
plt.plot(log_M_star_line, -7 + log_M_star_line, color='#D3D3D3', linestyle='-', label='sSFR = 100 Gyr$^{-1}$', zorder=3)
plt.plot(log_M_star_line, -8 + log_M_star_line, color='#C0C0C0', linestyle='-', label='sSFR = 10 Gyr$^{-1}$', zorder=3)
plt.plot(log_M_star_line, -9 + log_M_star_line, color='#A9A9A9', linestyle='-', label='sSFR = 1 Gyr$^{-1}$', zorder=3)
plt.plot(log_M_star_line, -10 + log_M_star_line, color='#696969', linestyle='-', label='sSFR = 0.1 Gyr$^{-1}$', zorder=3)

# Labeling the axes
plt.xlabel("$\\log_{10}(M_*/M_\\odot)$")
plt.ylabel("$\\log_{10}(SFR/M_\\odot \\ \\mathrm{yr^{-1}})$")
plt.legend(loc='upper left')

# Set plot limits
plt.ylim(-0.5, 4.5)
plt.xlim(8, 13)

plt.show()
```

```python
# Load the FLARES data from the CSV file
flares_data = pd.read_csv('../data/filtered_FLARES_data.csv')

# Extract Akins data and convert to numpy arrays
logMstar = np.array(akins_data["logMstar_med"][1:], dtype=float)
logMstar_upper68 = np.array(akins_data["logMstar_u68"][1:], dtype=float)
logMstar_lower68 = np.array(akins_data["logMstar_l68"][1:], dtype=float)
SFR_100 = np.log10(np.array(akins_data["SFR_100_med"][1:], dtype=float))
SFR_100_upper68 = np.log10(np.array(akins_data["SFR_100_u68"][1:], dtype=float))
SFR_100_lower68 = np.log10(np.array(akins_data["SFR_100_l68"][1:], dtype=float))

# Calculate uncertainties for stellar mass and SFR
logMstar_err = [logMstar - logMstar_lower68, logMstar_upper68 - logMstar]
SFR_100_err = [SFR_100 - SFR_100_lower68, SFR_100_upper68 - SFR_100]

# Extract FLARES data and filter for valid values
logMstar_flares = np.array(flares_data["Stellar Mass"], dtype=float)
SFR_100_flares = np.array(flares_data["Star Formation Rate"], dtype=float)
mask = (logMstar_flares > 0) & (logMstar_flares < 100) & (SFR_100_flares > 0)  # Filter for valid data
logMstar_flares_filtered = logMstar_flares[mask]
SFR_100_flares_filtered = np.log10(SFR_100_flares[mask])

# Create the figure with subplots
fig = plt.figure(figsize=(12, 11))

# Main scatter plot with error bars for Akins data
ax_scatter = fig.add_axes([0.1, 0.2, 0.65, 0.65])
subset_indices_akins = np.linspace(0, len(logMstar) - 1, num=400, dtype=int)  # Adjust number of points for clarity
logMstar_subset_akins = logMstar[subset_indices_akins]
SFR_100_subset_akins = SFR_100[subset_indices_akins]
logMstar_err_subset_akins = np.array([logMstar_err[0][subset_indices_akins], logMstar_err[1][subset_indices_akins]])
SFR_100_err_subset_akins = np.array([SFR_100_err[0][subset_indices_akins], SFR_100_err[1][subset_indices_akins]])

# Plot error bars
ax_scatter.errorbar(logMstar_subset_akins, SFR_100_subset_akins, 
                    xerr=logMstar_err_subset_akins, yerr=SFR_100_err_subset_akins, 
                    fmt='none', ecolor='orange', elinewidth=0.5, alpha=0.1, capthick=0)

# Hexbin plots for Akins and FLARES data
hb1 = ax_scatter.hexbin(logMstar, SFR_100, gridsize=50, cmap='Oranges', mincnt=1, zorder=6)
hb2 = ax_scatter.hexbin(logMstar_flares_filtered, SFR_100_flares_filtered, gridsize=50, cmap='Greens', mincnt=1)

# Plot specific star formation rate (sSFR) lines
log_M_star_line = np.linspace(0, 16, 100)
ax_scatter.plot(log_M_star_line, -10 + log_M_star_line, color='#5ec962', linestyle='-', alpha=0.4, label='sSFR = 0.1 Gyr$^{-1}$')
ax_scatter.plot(log_M_star_line, -8 + log_M_star_line, color='#21918c', linestyle='-', alpha=0.4, label='sSFR = 10 Gyr$^{-1}$')
ax_scatter.plot(log_M_star_line, -6 + log_M_star_line, color='#3b528b', linestyle='-', alpha=0.4, label='sSFR = 1000 Gyr$^{-1}$')

# Labeling the axes
ax_scatter.set_xlabel("$\\log_{10}(M_*/M_\\odot)$", fontsize=15)
ax_scatter.set_ylabel("$\\log_{10}(SFR/M_\\odot \\ \\mathrm{yr^{-1}})$", fontsize=15)
ax_scatter.legend(loc='lower right', fontsize=18)
ax_scatter.set_ylim(-2, 5)
ax_scatter.set_xlim(7.5, 13)

# Add histograms for Akins and FLARES data
ax_hist_x = fig.add_axes([0.1, 0.867, 0.65, 0.1], sharex=ax_scatter)
ax_hist_y = fig.add_axes([0.77, 0.2, 0.1, 0.65], sharey=ax_scatter)
ax_hist_x.hist(logMstar, bins=40, color="orange", alpha=0.3, density=True, edgecolor='black', linewidth=1.2)
ax_hist_x.hist(logMstar_flares_filtered, bins=15, color='green', alpha=0.3, density=True, edgecolor='black', linewidth=1.2)
ax_hist_y.hist(SFR_100, bins=40, color="orange", alpha=0.3, density=True, edgecolor='black', linewidth=1.2, orientation='horizontal')
ax_hist_y.hist(SFR_100_flares_filtered, bins=50, color='green', alpha=0.3, density=True, edgecolor='black', linewidth=1.2, orientation='horizontal')

# KDE plots for Akins and FLARES data
sns.kdeplot(logMstar, ax=ax_hist_x, color='orange')
sns.kdeplot(logMstar_flares_filtered, ax=ax_hist_x, color='green')
sns.kdeplot(SFR_100, ax=ax_hist_y, color='orange', vertical=True)
sns.kdeplot(SFR_100_flares_filtered, ax=ax_hist_y, color='green', vertical=True)

ax_hist_x.set_ylabel('Density', fontsize=12)
ax_hist_y.set_xlabel('Density', fontsize=12)

# Hide tick labels but keep tick marks for histograms
ax_hist_x.xaxis.set_tick_params(labelbottom=False)
ax_hist_y.yaxis.set_tick_params(labelleft=False)

# Add colorbars for hexbin plots
cb1 = fig.add_axes([0.1, 0.05, 0.65, 0.02])
cb2 = fig.add_axes([0.1, 0.127, 0.65, 0.02])
plt.colorbar(hb1, cax=cb1, orientation='horizontal', label='COSMOS-Web Galaxy Count')
plt.colorbar(hb2, cax=cb2, orientation='horizontal', label='FLARES Galaxy Count')

plt.show()
```
